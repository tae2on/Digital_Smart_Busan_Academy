```python
df.set_index('발전 전력별', inplace=True)
df = df.T
```

```python
# 증감율(변동률) 계산
df.rename(columns={'합계': '총발전량'}, inplace=True)
df['전년도총발전량'] = df['총발전량'].shift(1)
df['증감율'] = ((df['총발전량'] / df['전년도총발전량']) -1) * 100
```
