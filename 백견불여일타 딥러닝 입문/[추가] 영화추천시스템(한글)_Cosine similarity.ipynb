* tf(d,t) : 특정 문서 d에서의 특정 단어 t의 등장 횟수
* df(t) : 특정 단어 t가 등장한 문서의 수
* idf(d, t) : df(t)에 반비례하는 수



```python
import pandas as pd # 데이터프레임 사용을 위해
from math import log # IDF 계산을 위해

docs = [
    '먹고 싶은 사과',
    '먹고 싶은 바나나',
    '길고 노란 바나나 바나나',
    '저는 과일이 좋아요'
]
vocab = list(set(w for doc in docs for w in doc.split()))
```

```python
vocab
```

```python
N = len(docs) # 총 문서의 수

def tf(t, d):
  return d.count(t)
```

```python
result = []
for i in range(N): # 각 문서에 대해서 아래 명령을 수행
  result.append([])
  d = docs[i]
  for j in range(len(vocab)):
    t = vocab[j]
    result[-1].append(tf(t, d))
```

```python
tf_ = pd.DataFrame(result, columns = vocab)
tf_
```

```python
def idf(t):
  df = 0
  for doc in docs:
    df += t in doc
  return log(N / (df + 1))
```

```python
result = []
for j in range(len(vocab)):
  t = vocab[j]
  result.append(idf(t))

idf_ = pd.DataFrame(result, index = vocab, columns = ['IDF'])
idf_
```

```python
def tfidf(t, d):
  return tf(t, d) * idf(t)
```

```python
result = []
for i in range(N):
  result.append([])
  d = docs[i]
  for j in range(len(vocab)):
    t = vocab[j]
    result[-1].append(tfidf(t, d))

tfidf_ = pd.DataFrame(result, columns = vocab)
tfidf_
```

```python
morphs_lst = []
for i in tqdm(data_dropna.index):
  morphs = ''
  morph = okt.morphs(data_dropna.loc[i]['본문'], stem=True)
  for txt in morph:
    if txt not in stopword:
      morphs = morphs + ' ' + txt
  morphs_lst.append(morphs)

data_dropna['morphs'] = morphs_lst
data_dropna.head()
```

```python
# tf-idf: https://wikidocs.net/31698
tfidf = TfidfVectorizer()
# morphs 열에 대해서 tf-idf를 수행
tfidf_matrix = tfidf.fit_transform(data_dropna['morphs'])
print(tfidf_matrix.shape)
```

```python
# 코사인 유사도: https://wikidocs.net/24603
# 코사인 유사도를 사용하면 바로 문서의 ( )를 구한다.
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
cosine_sim
```
